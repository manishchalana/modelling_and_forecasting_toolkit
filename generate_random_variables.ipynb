## Import modules here

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Generate normal random variables

```python

def sample_normal_random_variables(n):
    arr = np.random.randn(n)
    print(f"Shape of array: {arr.shape}")
    fig, ax = plt.subplots(1, 2, figsize=(20, 12))
    sns.scatterplot(x=range(n), 
                    y=arr, 
                   ax=ax[0])
    ax[0].set_title("Scatterplot")
    sns.histplot(arr, ax=ax[1])
    ax[1].set_title("Histogram")
    fig.suptitle(f"{n} Normally distributed random variables")
    plt.show()
```

```python
num_points = 10
sample_normal_random_variables(n=num_points)
```

```python
num_points = 100
sample_normal_random_variables(n=num_points)
```

```python
num_points = 1000
sample_normal_random_variables(n=num_points)
```
